# Adding data to Axes object.
Adding data to a figure is done by calling methods of the Axes object.

```python
# This part creates figure and axes, generates data and plotes on axes.

import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()
x1 = np.array(np.random.randint(10, size=(10))) # x1 is array of length 10 containing random integers in range(0, 10)
y1 = np.array([x*2 for x in x1]) # y1 is array of length 10 containing x*2 elements.
ax.plot(x1, y1)
x2 = np.array(np.random.randint(10, size=(10))) # same as x1
y2 = np.array([x*3 for x in x2]) # same as x2
ax.plot(x2, y2)
plt.show()
```

# Customizing data appeareance
We can customize the appearance of data in our plots, while adding the data to the plot, using key-word arguments to the plot command.

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()
x1 = np.array(np.random.randint(10, size=(10))) # x1 is array of length 10 containing random integers in range(0, 10)
y1 = np.array([x*2 for x in x1]) # y1 is array of length 10 containing x*2 elements.
ax.plot(x1, y1, linestyle=":", marker="v", color='r') # Marker marks the points of x axis, color makes the line in red.
x2 = np.array(np.random.randint(10, size=(10))) # same as x1
y2 = np.array([x*3 for x in x2]) # same as x2
ax.plot(x2, y2, linestyle="-", marker='o', color='y') # Marker marks the points of x axis, color makes the line in yellow.
plt.show()
```

# Customizing axis labels and adding titles
Customizing the axis labels requires using the **set_xlabel** and **set_ylabel** methods of the Axes object. Adding a title uses the **set_title** method.

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()
x1 = np.array(np.random.randint(10, size=(10)))
y1 = np.array([x*2 for x in x1]) 
ax.plot(x1, y1, linestyle=":", marker="v", color='r') 
x2 = np.array(np.random.randint(10, size=(10)))
y2 = np.array([x*3 for x in x2]) 
ax.plot(x2, y2, linestyle="-", marker='o', color='y')
ax.set_xlabel("X values") # Setting the label of x-axis
ax.set_ylabel("Y values") # Setting the label of y-axis
ax.set_title("Some random plots") # Setting the title of the figure
plt.show()
```

# Creating small multiples with plt.subplots
Small multiples are used to plot several datasets side-by-side.  
In Matplotlib, small multiples can be created using the **plt.subplots()** function.  
The first argument is the number of rows in the array of Axes objects generate and the second argument is the number of columns.

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(2, 2)

x1 = np.array(np.random.randint(10, size=(10)))
y1 = np.array([x for x in x1]) 
ax[0, 0].plot(x1, y1, color='r')

x2 = np.array(np.random.randint(10, size=(10)))
y2 = np.array([x*2 for x in x2])
ax[0, 1].plot(x2, y2, color='g')

x3 = np.array(np.random.randint(10, size=(10)))
y3 = np.array([x*3 for x in x3]) 
ax[1, 0].plot(x3, y3, color='b') 

x4 = np.array(np.random.randint(10, size=(10)))
y4 = np.array([x*8 for x in x4])
ax[1, 1].plot(x4, y4, color='y')

plt.show()
```

# Small multiples with shared y axis
When creating small multiples, it is often preferable to make sure  
that the different plots are displayed with the same scale used on the y-axis.  
This can be configured by setting the sharey key-word to True.

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(2, 1, sharey=True)

x1 = np.array(np.random.randint(10, size=(10)))
y1 = np.array([x for x in x1]) 
ax[0].plot(x1, y1, color='r')

x2 = np.array(np.random.randint(10, size=(10)))
y2 = np.array([x*2 for x in x2])
ax[1].plot(x2, y2, color='g')
```
